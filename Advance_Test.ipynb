```python
import pandas as pd

# -------------------------------
# 1. Create Data
# -------------------------------
orders = pd.DataFrame({
    "order_id": [1,2,3,4,5,6,7,8],
    "customer_id": [101,102,101,103,104,102,105,101],
    "order_date": ["2024-01-05","2024-01-10","2024-02-15","2024-02-20",
                   "2024-03-05","2024-03-15","2024-03-20","2024-03-25"],
    "product_id": [11,12,13,11,12,13,12,11],
    "amount": [50000,2000,None,52000,3000,900,2500,51000]
})

customers = pd.DataFrame({
    "customer_id": [101,102,103,104,105],
    "name": ["A","B","C","D","E"],
    "city": ["Delhi","Mumbai","Delhi",None,"Mumbai"]
})

products = pd.DataFrame({
    "product_id": [11,12,13],
    "product": ["Laptop","Shirt","Mouse"],
    "category": ["Electronics","Clothing","Electronics"]
})

# -------------------------------
# 2. Data Cleaning
# -------------------------------
orders["order_date"] = pd.to_datetime(orders["order_date"])

# fill missing amount with median
orders["amount"] = orders["amount"].fillna(orders["amount"].median())

# fill missing city
customers["city"] = customers["city"].fillna("Unknown")

# -------------------------------
# 3. Merge Data
# -------------------------------
df = pd.merge(orders, customers, on="customer_id", how="inner")
df = pd.merge(df, products, on="product_id", how="inner")

# -------------------------------
# 4. Feature Engineering
# -------------------------------
df["year"] = df["order_date"].dt.year
df["month"] = df["order_date"].dt.month
df["day"] = df["order_date"].dt.day_name()

# -------------------------------
# 5. Filtering
# -------------------------------
avg_amount = df["amount"].mean()
filtered_df = df[(df["amount"] > avg_amount) & (df["city"] == "Delhi")]

# -------------------------------
# 6. Apply, Map, ApplyMap
# -------------------------------
df["value_type"] = df["amount"].apply(lambda x: "High Value" if x > 30000 else "Low Value")

df["sub_category"] = df["category"].map({
    "Electronics": "Tech",
    "Clothing": "Apparel"
})

# applymap (lowercase all strings)
df = df.applymap(lambda x: x.lower() if isinstance(x, str) else x)

# -------------------------------
# 7. GroupBy + Ranking
# -------------------------------
df["customer_rank"] = df.groupby("customer_id")["amount"].transform("sum").rank(ascending=False, method="dense")


top_customer_per_city = df.loc[df.groupby("city")["amount"].idxmax()]

# -------------------------------
# 8. Pivot Table
# -------------------------------
pivot_table = df.pivot_table(
    values="amount",
    index="month",
    columns="category",
    aggfunc="sum"
)

# -------------------------------
# 9. Sorting
# -------------------------------
sorted_df = df.sort_values(["city", "amount"], ascending=[True, False])

# -------------------------------
# 10. String Operations
# -------------------------------
df["name"] = df["name"].str.upper()
df["product_first_letter"] = df["product"].str[0]

# -------------------------------
# 11. Transform Analysis
# -------------------------------
df["customer_avg"] = df.groupby("customer_id")["amount"].transform("mean")
df["category_avg"] = df.groupby("category")["amount"].transform("mean")

df["compare_with_cusavg"] = df["amount"] - df["customer_avg"]
df["compare_with_catavg"] = df["amount"] - df["category_avg"]

# -------------------------------
# 12. Insights
# -------------------------------
city_revenue = df.groupby("city")["amount"].sum().sort_values(ascending=False)

category_trend = df.pivot_table(values="amount", index="month", columns="category", aggfunc="sum")

day_sales = df.groupby("day")["amount"].sum().sort_values(ascending=False)

customer_consistency = df.groupby("customer_id")["amount"].std().sort_values()

# -------------------------------
# 13. Export
# -------------------------------
df.to_csv("dataset.csv", index=False)

from google.colab import files
files.download("dataset.csv")
```
